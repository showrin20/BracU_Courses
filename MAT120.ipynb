```python
from sympy import*
x=symbols("x")
fi= exp(-x)*(x**2)*cos(x)
integrate(fi, x)
```

```python
### PRAC
fi=((sin(x))**3)*exp(-5*x)
integrate(fi,x)
```

```python
## prac simply
fi=((sin(x))**3)*exp(-5*x)
integrate(fi,x).simplify()
```

```python
### Prac 2

fi=((1+sqrt(x))**(1/3))/sqrt(x)
integrate(fi, x)
```

```python
## Prac update

fi=((1+sqrt(x))**Rational(1,3))/sqrt(x)
integrate(fi, x).simplify()
```

```python
### LIMIT

fi=exp(x)/(sqrt(exp(x)+9))
integrate(fi, (x,0,log(4))).evalf()
```

```python
### frst derivative

from sympy import*
init_printing()
x=symbols("x")
y=Function("y")
y1=Derivative(y(x),x)
eqq=y1+(y(x)**2)/x**2
sol=dsolve(eqq,y(x),ics={y(1):"1/2"})
pprint(sol)
```

```python
from sympy.plotting import*
p1=plot(sol.rhs,(x,-1,1/3-(1e-3)),ylim=[-10,10])
p2=plot(sol.rhs,(x,1/3+(1e-4),1),ylim=[-10,10])
p1.extend(p2)
p1.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Define x using numpy to avoid division by zero
x = np.linspace(-10, 10, 1000)  # Dense range of x values avoiding zero
x = x[x != 0]  # Exclude zero from x

# Differentiation 1
y1 = 32 * x**15 - 8 * x + (-3 / x**4)

# Plotting Differentiation 1
plt.figure(figsize=(10, 6))
plt.plot(x, y1, label="y = 32*x^15 - 8*x + (-3/x^4)")
plt.xlabel("x - axis")
plt.ylabel("y - axis")
plt.title("Differentiation 1")
plt.legend()
plt.grid()
plt.show()

# Differentiation 2
y2 = 480 * x**14 - 8 - (-12 / x**5)

# Plotting Differentiation 2
plt.figure(figsize=(10, 6))
plt.plot(x, y2, label="y = 480*x^14 - 8 - (-12/x^5)", color="red")
plt.xlabel("x - axis")
plt.ylabel("y - axis")
plt.title("Differentiation 2")
plt.legend()
plt.grid()
plt.show()
```
